# Introduction
For the project, we gathered the dataset _Maryland Licensed Animal Plant Facilities - Licensed Plant Facilities_ from the official Maryland government website to perform data cleaning. The dataset contains licensed Plant Brokers, Dealers and Nurseries in Maryland. Attributes include location and contact information.

https://data.imap.maryland.gov/datasets/maryland-licensed-animal-plant-facilities-licensed-plant-facilities

The dataset is of 1906rows, 17columns and contains some of the data in coded format(PRODUCT_TYPE, BUS_CODE, BUS_STATUS, W_R) and their descriptions are found in the following link: 

https://mda.maryland.gov/plants-pests/Pages/nurseries_plant_dealers.aspx

In order to clean and visualize the dataset we used pandas, matplotlib and seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plants = pd.read_csv('Maryland_Licensed_Animal_Plant_Facilities_-_Licensed_Plant_Facilities.csv')
plants

# Reading the csv file
```

```python
plants.info() 

# Information of the plants dataframe
```

```python
plants.isnull().sum()

# checking for null values in the dataset
```

```python
plants[plants.PHONE1 == plants.PHONE2]

# checked for columns that contain both the phone1 and phone2 with same phonenumbers which are duplicates
```

```python
a=plants[plants.PHONE1.isnull() & plants.PHONE2]
a
# checked for rows that cointain phonenumber in phone2 column and a NULL value in phone1 column. 
```

```python
a["PHONE1"].fillna(a["PHONE2"], inplace=True)

# filled phone1 column null values with phone2 phonenumbers
```

```python
plants[plants.PHONE1.isnull()]
# Now the null values in phone1 column got decreased 
```

```python
plants.pop('PHONE2')
plants.rename(columns = {'PHONE1':'PhoneNumber'}, inplace = True)
# After doing necessary changes popped the phone2 column as there are mostly null values and duplicate values in it.
# Renamed the column Phone1 to PhoneNumber
```

```python
plants[['product_type1', 'product_type2', 'product_type3', 'product_type4', 'product_type5', 'product_type6', 'product_type7', 'product_type8', 'product_type9', 'product_type10', 'product_type11', 'product_type12', 'product_type13', 'product_type14']] = plants['PRODUCT_TYPE'].str.split(',', expand=True)

plants.pop('PRODUCT_TYPE')

plants

# As there are multiple values in PRODUCT_TYPE field, we splitted the PRODUCT_TYPE column into multiple columns with single value in each field.
# Popped the PRODUCT_TYPE column after splitting it.
```

```python
plants[['Business_type1', 'Business_type2', 'Business_type3', 'Business_type4','Business_type5']] = plants['BUS_CODE'].str.split(',', expand=True)
plants.pop('BUS_CODE')
plants
```

```python
plants[['Sales_type1', 'Sales_type2', 'Sales_type3', 'Sales_type4']] = plants['W_R'].str.split(',', expand=True)
plants.pop('W_R')
plants
```

```python
# number of PLANTS for each different type of LICENSE
plants_licencetypes = plants.LICTYPE.value_counts()
plants_licencetypes
```

```python
plot = sns.catplot(data = plants, kind = 'count', x = 'LICTYPE')
plot.ax.set_title('Number of PLANTS for different type of LICENSES')

# Plotting Categorical Variables (Licence types)
# Used catplot to count the plants associated with different license types 
# We can also include the categories we want and the way they are to be ordered by using an order set.
```

```python
plot = sns.displot(data = plants, x= 'LIC_NO', kind = 'hist', height=3 ,aspect=2)
plot.ax.grid(axis = 'x')
plot.ax.set_title('Distribution of LICENSE NUMBERS')

# Plotting Continuous Variables (License Numbers)
# Used displot to see the distribution of LIC_NO column
```

```python
plot = sns.displot(plants, x="COUNTY", y="LICTYPE", cbar = True)
plot.ax.tick_params(labelrotation = 90)
plot.ax.set_title('Distribution between Counties and License types')

# Plotting Categorical Variables vs Categorical Variables (Counties vs license types)
# gaining insights between categorical vaiables is easier with displot.
```

```python
plot=sns.displot(data=plants, x="LICTYPE", kind="hist", hue = 'COUNTY')
plot.ax.set_title('Count of plants using Counties and License types')
```

```python
plot= sns.displot(data=plants, kind="ecdf", x="OBJECTID", hue = 'LICTYPE', aspect=2, alpha=0.7)
plot.ax.grid()
plot.ax.set_title('Distribution between OBJECTID and License type')
# Categorical vs Numerical varibles (objectID vs license type)
# Here Empirical cumulative distribution function(ecdf) uses the 'OBJECTID' and 'LICTYPE' columns
# present the proportion of observations that falls under each unique value in a dataset without any adjustments.
# hue parameter maps the license type categories with different colors.
```

```python
plot = sns.displot(data = plants, x="OBJECTID", y="LIC_NO", height= 8, aspect= 1.5)
plot.ax.set_title('Distribution between OBJECTID and License number')
# plotting Continuous vs Continuous Variables (ObjectID vs License Number)
```
