````python
import random 
import matplotlib.pyplot as plt

trial_count = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    """
    
    options = [1,2,3,4,5,6]
    simulations = [random.choice(options) for _ in range(trials)]
    return(simulations)
fair_die_simulation(trial_count)
````

````python
def biased_die_simuations(trials, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15]):
    """
    returns an array of randomly numbers between 1 and 6 with the probabilty of assosiated weight. 
    for instance: for weights: [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]
    1 has 0.25 change to be selected, 2 has 0.15 chance and so on...
    """
    
    options = [1,2,3,4,5,6]
    simulations = [random.choices(options, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15])[0] for _ in range(trials) ]
    return(simulations)
biased_die_simuations(trial_count)
````

````python
biased = biased_die_simuations(trial_count)
fair = fair_die_simulation(trial_count)
print("biased_mean_list:" + str(biased))
print("unbiased_mean_list:" + str(fair))

# biased_mean_list = ...
# unbiased_mean_list = ...
````

# Create 2 visualizations that shows outcomes of multiple rolls of a fair die and biased die.

## Visualization for fair die


````python
import pandas
def show_fairdie_roll_counts_df(trial_count):
    fair = fair_die_simulation(trial_count)
    roll_1 = fair.count(1)
    roll_2 = fair.count(2)
    roll_3 = fair.count(3)
    roll_4 = fair.count(4)
    roll_5 = fair.count(5)
    roll_6 = fair.count(6)
    
    df = pandas.DataFrame([{'No. of 1s':roll_1,'No. of 2s':roll_2, 'No. of 3s':roll_3,'No. of 4s':roll_4, 'No. of 5s':roll_5, 'No. of 6s':roll_6}])
    print(df)
    
    _=df.plot.bar() 
    _=plt.xticks([])
````

````python
show_fairdie_roll_counts_df(500)
````

# Repeat the process programatically and take the mean to create a new list. This list will be used to compare means of biased & unbiased dice.
## Create a historgram showing means of biased & unbiased die

````python
import numpy as np 
list_of_trial_results = []
number_of_fair_rolls_per_trial = 10    
number_of_trials = 10
for trial_indx in range(number_of_trials):
    options = [1,2,3,4,5,6]
    simulations = [random.choice(options) for _ in range(number_of_fair_rolls_per_trial)]
    list_of_trial_results.append(simulations)
    result = [np.mean(k) for k in zip(*list_of_trial_results)]

plt.xlabel('mean of fair dice')
plt.ylabel('number of trials')
plt.grid(True)
plt.title('Unbiased_histogram')
_=plt.hist(list_of_trial_results,bins=range(number_of_fair_rolls_per_trial))
plt.show()
````

````python
list_of_btrial_results = []
number_of_biased_rolls_per_trial = 10    
number_of_trials = 10
for trial_indx in range(number_of_trials):     
    options = [1,2,3,4,5,6]
    simulations = [random.choices(options, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15])[0] for _ in range(number_of_biased_rolls_per_trial) ]
    list_of_btrial_results.append(simulations)
    result = [np.mean(k) for k in zip(*list_of_btrial_results)]


plt.xlabel('mean of biased dice')
plt.ylabel('number of trials')
plt.grid(True)
plt.title('Biased_histogram')
_=plt.hist(list_of_trial_results, bins=range(number_of_biased_rolls_per_trial))
plt.show()
````
